```python
import requests
import xml.etree.ElementTree as ET
import os
import time
import re

xml_file = 'arxiv_cs_AI_recent.xml'  # If you have an XML file

try:
    tree = ET.parse(xml_file)
    root = tree.getroot()
except ET.ParseError as e:
    print(f'Error parsing XML: {e}')
    exit(1)

namespaces = {
    'arxiv': 'http://arxiv.org/schemas/atom',
    'dc': 'http://purl.org/dc/elements/1.1/',
    'atom': 'http://www.w3.org/2005/Atom',
    'content': 'http://purl.org/rss/1.0/modules/content/'
}

pdf_dir = 'arxiv_pdfs_names'
os.makedirs(pdf_dir, exist_ok=True)

def sanitize_filename(filename):
    filename = re.sub(r'[<>:"/\\|?*]', '', filename)
    filename = filename.replace(' ', '_')
    return filename[:100].strip()

for item in root.findall('channel/item'):
    link = item.find('link').text
    title = item.find('title').text

    filename = sanitize_filename(title) + '.pdf'
    pdf_path = os.path.join(pdf_dir, filename)

    if os.path.exists(pdf_path):
        print(f'PDF "{filename}" already exists. Skipping download.')
        continue

    arxiv_id = link.split('/')[-1]
    pdf_url = f'https://arxiv.org/pdf/{arxiv_id}.pdf'

    print(f'Downloading PDF for "{title}" from {pdf_url}...')
    try:
        response = requests.get(pdf_url, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f'Error downloading "{title}": {e}')
        continue

    with open(pdf_path, 'wb') as f:
        f.write(response.content)
    print(f'Saved PDF to "{pdf_path}"')

    time.sleep(5)
```
